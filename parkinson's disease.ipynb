#load

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix, roc_auc_score
)
```

# Load data anlysis

```python
data = pd.read_csv('/content/parkinsons.csv')
data.head()
```

```python
data.shape
data.info()
data.isnull().sum()
data.describe()
```

# target variable


```python
data = data.dropna()
print('After dropna',data.shape)
```

```python
target = 'status'
X = data.drop(columns=target)
y = data[target]
print("Features shape:", X.shape, " Target distribution:\n", y.value_counts())
```

```python
#grouping data based on target varianle
data.groupby('status').mean(numeric_only=True)
```

```python
#separating feature & target
X = data.drop(['name', 'status'], axis=1)
y = data['status']
print(X)
```

```python
print(y)
```

# preprocess

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

#trrain/test

```python
X_train,X_test,y_train,y_test = train_test_split(
    X_scaled,y,test_size=0.2,random_state=42
)
print(X.shape,'train size', X_train.shape, 'test size', X_test.shape)
```

# standardization

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)

X_test = scaler.transform(X_test)
```

#models

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "SVM (RBF Kernel)": SVC(kernel='rbf', probability=True),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42)
}
for name, mdl in models.items():
    mdl.fit(X_train, y_train)
    print(f"{name} trained")
```

#evaluation test

```python
results = {}
for name, mdl in models.items():
    X_eval = X_test
    y_pred = mdl.predict(X_eval)
    y_prob = mdl.predict_proba(X_eval)[:,1]

    results[name] = {
        'Accuracy': accuracy_score(y_test, y_pred),

        'ROC-AUC': roc_auc_score(y_test, y_prob)
    }
    print(f"\n{name} Classification Report:")
    print(classification_report(y_test, y_pred))

# display summary table
pd.DataFrame(results).T
```

#cross validation

```python
# Define the scoring metrics you want to use
scoring = ['accuracy', 'roc_auc']

# Initialize a dictionary to store cross-validation results
cv_metrics = {}

y_res = y_train

for name, mdl in models.items():
    # Calculate cross-validation scores for each metric
    scores = {
        metric: cross_val_score(mdl, X_train, y_res,
                                 cv=5, scoring=metric).mean()
        for metric in scoring
    }
    # Store the mean scores for the current model
    cv_metrics[name] = scores

# Display the cross-validation results
pd.DataFrame(cv_metrics).T
```
